```python
import numpy as np

print("--- EXERCISE 4: NUMPY ARRAYS ---")

# 1. Create a 1D array containing numbers from 10 to 19
array_1d = np.arange(10, 20)
print("1D Array:")
print(array_1d)

# 2. Reshape it into a 2x5 2D matrix
matrix_2d = array_1d.reshape(2, 5)
print("\nReshaped 2x5 Matrix:")
print(matrix_2d)

# 3. Perform elemental math: Multiply every element by 2
multiplied_matrix = matrix_2d * 2
print("\nMatrix Elements Multiplied by 2:")
print(multiplied_matrix)
```

```python
import pandas as pd

print("--- EXERCISE 5: PANDAS DATAFRAMES ---")

# 1. Create a dictionary with sample student data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [21, 22, 20, 23],
    'Grade': [85, 92, 78, 90]
}

# 2. Convert the dictionary into a Pandas DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# 3. Filter the DataFrame to show only students older than 21
filtered_df = df[df['Age'] > 21]
print("\nStudents Older Than 21:")
print(filtered_df)

# 4. Calculate and print the mean (average) grade of all students
mean_grade = df['Grade'].mean()
print(f"\nAverage Grade of all students: {mean_grade}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print("--- EXERCISE 6: DATA VISUALIZATION ---")

# 1. Generate 100 random data points following a normal distribution
# Setting a seed ensures the "randomness" is identical every time it runs
np.random.seed(42)
data_points = np.random.normal(loc=50, scale=10, size=100)

# 2. Initialize the plotting style and create a figure context
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))

# 3. Create a histogram overlaid with a Kernel Density Estimate (KDE) line
sns.histplot(data_points, kde=True, color="skyblue", bins=15)

# 4. Add titles and axis labels
plt.title("Normal Distribution Histogram (Lab 0 Practice)", fontsize=14, pad=15)
plt.xlabel("Values", fontsize=12)
plt.ylabel("Frequency Count", fontsize=12)

# 5. Display the final rendered graphic element inside the notebook
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

print("--- EXERCISE 7: LINEAR REGRESSION MODEL ---")

# 1. Create a toy dataset representing a simple trend
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([12, 19, 31, 38, 51])

# 2. Instantiate and train (fit) the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# 3. Use the trained model to make a prediction for a new house size
new_size = np.array([[6]])
predicted_price = model.predict(new_size)

print(f"Trained model weights (Slope/Coefficient): {model.coef_[0]:.2f}")
print(f"Trained model bias (Intercept): {model.intercept_:.2f}")
print(f"\nPrediction: A house size of 6 will cost approximately: ${predicted_price[0]:.2f}K")
```
